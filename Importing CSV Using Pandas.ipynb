## You can read in a CSV file to make an index with `index_col= "Column_name"`

```python
df1 = pd.read_csv("IMDB-Movie-Data.csv", index_col="Title")
```

### Finding the count for all of the Null values

```python
df1.isnull().sum()
```

### Replace all of the Null values in the Revenue column with the mean Revenue

```python
df1["Revenue (Millions)"].fillna(df1["Revenue (Millions)"].mean(), inplace=True)
```

### And since the "Metascore" column also had Null values we can also replace those, just as we did above.

```python
df1["Metascore"].fillna(df1["Metascore"].mean(), inplace=True)
```

### Running the `isnull().sum()` you can now see there null values are not present

```python
df1.isnull().sum()
```

### Or drop the whole Rank column entirely, while replacing the whitespace with `inplace=True`

```python
df1.drop(["Rank"], axis=1, inplace=True)
```

### You can group aspects by column with `groupby`

```python
df1.groupby(["Year"]).mean()
```

### We find the top grossing movies by year

```python
df1.groupby(["Year"]).mean().sort_values(by=["Revenue (Millions)"], ascending=False)
```
